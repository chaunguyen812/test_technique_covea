```python
# Créer une fonction qui détecte les valeurs outliers par la règle de quantile Q1 et Q3
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
# Remove les valeurs outliers dans la variable Nombre de pieces principales #
df_fn = remove_outlier(df_fn,"Nombre pieces principales")
```

# Transformer les variables catérogies #

```python
# Nombre des catérogies de la variable de code departement #
len(df_fn["Code departement"].value_counts())
```

```python
# Jointure le nom des régions # 
region = pd.read_csv("code-officiel-geographique-2019-regions-et-departement.csv",sep=";")
```

```python
region.head()
```

```python
df_fn.head()
```

```python
# Créer une fonction qui permet de corriger de code de département # 
def correct_code_depart(x):
    if x==1:
        x="01"
    elif x==2:
        x="02"
    elif x==3:
        x="03"
    elif x==4:
        x="04"
    elif x==5:
        x="05"
    elif x==6:
        x="06"
    elif x==7:
        x="07"
    elif x==8:
        x="08"
    elif x==9:
        x="09"
    else:
        x=str(x)
    return(x)
# Apply fonction pour corriger le code de département #
df_fn["Code departement"] = df_fn["Code departement"].apply(correct_code_depart)
```

```python
# Boucle pour jointure de région en fonction le code de département # 
reg = []
for i in df_fn["Code departement"].tolist():
    if i in region["Code INSEE Département"].tolist():
        name = region[region["Code INSEE Département"]==i]["Nom Région majuscule"].tolist()[0]
    else:
        name ="NA"
    reg.append(name)
# Créer la nouvelle variable de region #
df_fn["Region"] = reg
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
# Evaluer la qualité de modèle #
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(metrics.mean_squared_error(y_test, predictions))
r2 = metrics.r2_score(y_test, predictions)

print('MAE (Mean Absolute Error): %s' %mae)
print('MSE (Mean Squared Error): %s' %mse)
print('RMSE (Root mean squared error): %s' %rmse)
print('R2 score: %s' %r2)
```
